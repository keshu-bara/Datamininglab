```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("medical.csv")
```

```python
df = pd.DataFrame(data)
```

```python
# Select features for clustering and handle missing values
df_cluster = df[['Age', 'Cholesterol', 'Heart_Rate']].dropna()
```

```python
# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cluster)
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Print clustered data
print("Clustered Data:\n")
print(df_cluster)
```

```python
# Optional: Visualize the clusters
sns.scatterplot(x='Age', y='Cholesterol', hue='Cluster', data=df_cluster, palette='viridis')
plt.title("K-Means Clustering of Patients")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```
